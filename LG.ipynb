### Generate the log files per windows  

```python
for co in col :
    tmp = data[["Unnamed: 0",str(co)]]
    tmp2 = tmp.loc[data[str(co)] != 0] #enlever les lignes avec des zéros
    tmp3 = tmp2.loc[data["Unnamed: 0"] != "**********************************************"]
    st = tmp3.to_string(index=False, header=False, justify="start")
    with open ("C:/Users/opale/Desktop/stage/supertri_56_2000/"+str(co)+".log", "w") as f : # TO DO : insert the save emplacment
        f.write(st)
```

### Change the - in . 

```python
for co in col :
    with open("C:/Users/opale/Desktop/stage/supertri_56_2000/"+str(co)+".log", 'r') as fin: # TO DO : insert the path of the log file
        lines = fin.readlines()
        del lines[0]
    with open("C:/Users/opale/Desktop/stage/supertri_56_2000/"+str(co)+".log", 'w') as fout: # TO DO : insert the path of the log file
        for line in lines:
            fout.write(line.replace('-', '.'))
```

### Remove empty column in the left 

```python
for co in col :
    with open("C:/Users/opale/Desktop/stage/supertri_56_2000/"+str(co)+".log", 'r') as fin: # TO DO : insert the path of the log file
        lines = fin.readlines() 
    with open("C:/Users/opale/Desktop/stage/supertri_56_2000/"+str(co)+".log", 'w') as fout: # TO DO : insert the path of the log file
        for line in lines:
            fout.write(line[1:])
```
