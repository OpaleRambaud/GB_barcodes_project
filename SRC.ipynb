```python
def get_names_from_dict(dic_name_int, sequence):
    """
    split string and retrieves a list of strings
    """
    
    for value in dic_name_int[sequence]:
        names = value.split(" + ")
    
    return names
```

```python
def get_names_from_row(row):
    """
    retrieves the row
    """
    
    return data2["Names"][row].split(" + ")
```

```python
lst_contain = {}
for row in rows2:
    row_name_list = get_names_from_row(row)
    #print(row_name_list)
    for seq, names in dic_name_int.items():
        if seq not in lst_contain.keys():
            lst_contain[seq] = []
        check_name = set(names) & set(row_name_list)
        if check_name:
            if row not in lst_contain[seq]:
                lst_contain[seq].append(row)
```

### Find the fragment with low signal

```python

lim_bas = 30
lim_haut = 50 


col_bas2 = {}
cmpt_bas = 0
nb_bas2 = {}
ideb2 = {}


for seq in val :
    
    col_bas2[seq] = []
    nb_bas2[seq] = []
    ideb2[seq] = []

    
    for column in data2.columns[2:] :

        if data2[column][seq] < lim_bas  and cmpt_bas == 0:


            idebut_bas= int(column)

            cmpt_bas += 1

        if data2[column][seq] < lim_bas and cmpt_bas >= 1: 

            cmpt_bas += 1 

        if data2[column][seq] < lim_bas : 
            
            nb_bas2[seq].append(column)
            
            
        if column == "557" and data2["557"][seq] < lim_bas and cmpt_bas >= 1: # TO DO : insert the number of windows
            
            ifin_bas = int(column)

            col_bas2[seq].append((idebut_bas,ifin_bas))

            ideb2[seq].append(idebut_bas)
            
            cmpt_bas = 0 


        if data2[column][seq] >= lim_bas and cmpt_bas >= 1:

            ifin_bas = int(column) -1

            col_bas2[seq].append((idebut_bas,ifin_bas))

            ideb2[seq].append(idebut_bas)

            cmpt_bas = 0
```

### Find the fragments with hight signal 

```python

col_haut = {}
cmpt_haut = 0
nb_haut = {}

for seq in rows2 :
    
    col_haut[seq] = []
    nb_haut[seq] = []
        
    for column in data2.columns[2:] :

        if data2[column][seq] > lim_haut  and cmpt_haut == 0:

            i5prime_haut= int(column)

            cmpt_haut += 1

        if data2[column][seq] > lim_haut and cmpt_haut >= 1: 

            cmpt_haut += 1 
                        
        if data2[column][seq] > lim_haut :  
            
            nb_haut[seq].append(column)
            
         
        if column == "557" and data2["557"][seq] > lim_haut and cmpt_haut >= 1: # TO DO : insert the number of windows
            
            i3prime_haut = int(column)
            col_haut[seq].append((i5prime_haut,i3prime_haut))
            
            cmpt_haut = 0
            
        if data2[column][seq] <= lim_haut and cmpt_haut >= 1:

            i3prime_haut = int(column) -1
            col_haut[seq].append((i5prime_haut,i3prime_haut))
            cmpt_haut = 0
```

### Find the candidates 

```python
dic_cand = {}
l_ta = []
moy_boot = {}
moy_boot2 = {}
for seq in val :
    print("ligne d'interet :")
    print(seq)
    name = data2["Names"][seq]
    name2 = data2["Names"][seq].split(" + ")
    cor_name = []
    for n in name2 :
        cor_name.append(dic[n])
    cor_name2 = " + ".join(cor_name)
        
    for i in col_bas2[seq]:

        idebut= i[0]
        coor = win_nucl[idebut-1][0]

        ifin = i[1]
        coor2 = win_nucl[ifin-1][1]
        if ifin != 557 : # TO DO : insert the number of windows
            coor3 = win_nucl[ifin][0]
        else :
            coor3 = win_nucl[ifin-1][1]

        tai = ifin - idebut +1 

    
        for row in lst_contain[seq] :


            name_r = data2["Names"][row]
            name_r2 = data2["Names"][row].split(" + ")
            cor_name_r = []
            cmpt = 0
            cond = 0
            for nr in name_r2 :
                cor_name_r.append(dic[nr])
            cor_name_r2 = " + ".join(cor_name_r)
            for nrr in name_r2 :
                if nrr not in name2 :
                    cond = 1
                if nrr in name2 : 
                    cmpt += 1
            if cmpt < len(name2) and cond == 1:
                
                for j in col_haut[row] :
                    idebut2 = j[0]

                    ifin2 = j[1]

                    ta = (ifin2 - idebut2) +1
                    t = (coor3 - coor) +1

                    if int(idebut2) == int(idebut) :
                        if int(ifin2) == int(ifin)  :
                            cmpt2 =0
                            for sign in range(idebut2,ifin2 +1) :

                                if int(data2[str(sign)][row]) + int(data2[str(sign)][seq]) <= 100:
                                    cmpt2 += 1
                            if cmpt2 == ta :
                                
                                if t >= 500 :
                                    val_col = []
                                    val_col2 = []
                                    
                                    for col in range(1,558) :
                                        col_st = str(col)
                                        boot = int(data2[col_st][seq])
                                        val_col.append(boot)
                                        boot2 = int(data2[col_st][row])
                                        val_col2.append(boot2)
                                    moy_boot[cor_name2] = np.mean(val_col)
                                    moy_boot2[cor_name_r2] = np.mean(val_col2)
                                
                                    l_ta.append(t)


                                    print("Ligne concernée :")
                                    print(cor_name2)
                                    print("Candidat :")
                                    print(cor_name_r2)
                                    print("Fragments concerné :")
                                    print(coor,coor3)
                                    print("*******************")

                                    if cor_name2 not in dic_cand :
                                        dic_cand[cor_name2] = {}

                                    if (coor,coor3) not in dic_cand[cor_name2] :

                                        dic_cand[cor_name2][(coor,coor3)] = []

                                        dic_cand[cor_name2][(coor,coor3)].append(cor_name_r2)

                                    else :

                                        dic_cand[cor_name2][(coor,coor3)].append(cor_name_r2)




```
